```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
from sklearn.preprocessing import MinMaxScaler
```

```python
X= -2 * np.random.rand(100,2)
```

```python
X
```

A total of 100 data points has been generated and divided into two groups, of 50 points each.

```python
X1 = 1 + 2 * np.random.rand(50,2)
```

```python
X1
```

```python
X[50:100, :] = X1
```

```python
X1
```

```python
plt.scatter(X[ : , 0], X[ :, 1], s = 50)
```

```python
from sklearn.cluster import KMeans
Kmean = KMeans(n_clusters=2)
Kmean.fit(X)
```

```python
Kmean.cluster_centers_
```

```python
plt.scatter(X[ : , 0], X[ : , 1], s =50)

plt.scatter(-0.94665068, -0.97138368, s=200)

plt.scatter(2.01559419, 2.02597093, s=200)

plt.show()
```

50 data points belong to the 0 cluster while the rest belong to the 1 cluster.

```python
Kmean.labels_
```

It shows that the test data point belongs to the 0 (green centroid) cluster.

```python
sample_test=np.array([-3.0,-3.0])
second_test=sample_test.reshape(1, -1)
Kmean.predict(second_test)
```
