```python
import pandas as pd
import numpy as np
```

```python
#to read the data in the csv file
data = pd.read_csv("/content/Dhoni_Last_10_Match_Stumping_History.csv")
print(data,"n")
```

```python
#making an array of all the attributes
d = np.array(data)[:,:-1]
print("n The attributes are: ",d)
```

```python
#segragating the target that has positive and negative examples
target = np.array(data)[:,-1]
print("n The target is: ",target)
```

```python
#training function to implement find-s algorithm
def train(data_records,target):
    for index, val in enumerate(target):
        if val == "Yes":
            specific_hypothesis = data_records[index].copy()
            break
             
    for index, val in enumerate(data_records):
        if target[index] == "Yes":
            for x in range(len(specific_hypothesis)):
                if val[x] != specific_hypothesis[x]:
                    specific_hypothesis[x] = '?'
                else:
                    pass
                 
    return specific_hypothesis
 
#obtaining the final hypothesis
print("n The final hypothesis is:",train(d,target))
```
